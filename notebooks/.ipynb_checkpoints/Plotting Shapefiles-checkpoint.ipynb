```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib as mpl

from matplotlib.colors import LinearSegmentedColormap
```

```python
%matplotlib inline

sns.set(style='darkgrid', palette='pastel', color_codes=True)
sns.mpl.rc('figure', figsize=(10,6))
```

```python
# Using the 110th congress as a trial run
shapefiles_path = '../congress-data/shapefiles/districtShapes110/districts110.shp'
sf = shp.Reader(shapefiles_path)
```

```python
def read_shapefile(sf):
    ## Code taken from https://towardsdatascience.com/mapping-geograph-data-in-python-610a963d2d7f
    """
    Read a shapefile into a Pandas dataframe with a 'coords' 
    column holding the geometry information. This uses the pyshp
    package
    """
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shps = [s.points for s in sf.shapes()]
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    return df
```

```python
def map_colors(party_list, pct_list):
    '''
        Maps a color to a specific political party and percentage of votes missed.
        party_list: array of strings
        pct_list: array of percentages
        
        Returns: array of hex codes
    '''
    
    fill_cols = []
    for i in range(len(party_list)):
        
        party = party_list[i]
        pct = pct_list[i]
        
        if party == 'R':
            if pct <= 0.01:
                fill_cols.append('#FFEBEE')
            elif pct <= 0.03:
                fill_cols.append('#FFCDD2')
            elif pct <= 0.05:
                fill_cols.append('#EF9A9A')
            elif pct <= 0.07:
                fill_cols.append('#E57373')
            elif pct <= 0.09:
                fill_cols.append('#EF5350')
            elif pct <= 0.15:
                fill_cols.append('#F44336')
            else:
                fill_cols.append('#E53935')
            
        elif party == 'D':
            if pct <= 0.01:
                fill_cols.append('#E3F2FD')
            elif pct <= 0.03:
                fill_cols.append('#BBDEFB')
            elif pct <= 0.05:
                fill_cols.append('#90CAF9')
            elif pct <= 0.07:
                fill_cols.append('#64B5F6')
            elif pct <= 0.09:
                fill_cols.append('#42A5F5')
            elif pct <= 0.15:
                fill_cols.append('#2196F3')
            else:
                fill_cols.append('#1E88E5')
                
        elif party == 'O1':
            if pct <= 0.01:
                fill_cols.append('#F1F8E9')
            elif pct <= 0.03:
                fill_cols.append('#DCEDC8')
            elif pct <= 0.05:
                fill_cols.append('#C5E1A5')
            elif pct <= 0.07:
                fill_cols.append('#AED581')
            elif pct <= 0.09:
                fill_cols.append('#9CCC65')
            elif pct <= 0.15:
                fill_cols.append('#8BC34A')
            else:
                fill_cols.append('#7CB342')
                
        elif party == 'O2':
            if pct <= 0.01:
                fill_cols.append('#FFFDE7')
            elif pct <= 0.03:
                fill_cols.append('#FFF9C4')
            elif pct <= 0.05:
                fill_cols.append('#FFF59D')
            elif pct <= 0.07:
                fill_cols.append('#FFF176')
            elif pct <= 0.09:
                fill_cols.append('#FFEE58')
            elif pct <= 0.15:
                fill_cols.append('#FFEB3B')
            else:
                fill_cols.append('#FDD835')
            
        else:
            raise ValueError('Unknown political party identifier: %s' % party)
            
    return fill_cols
```

```python
def plot_single_shape(sf, dist_id, fill_col, figsize=(10, 6)):
    '''
        Plots a single shape.
        Takes a shapefile and the id of the district in the shapefile
        
        Returns the mean center of the shape
    '''
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_aspect('equal')
    shape_obj = sf.shape(dist_id)
    ## Plot longs on the x axis, lats on the y axis
    longs = np.zeros((len(shape_obj.points), 1))
    lats = np.zeros((len(shape_obj.points),1))
    
    for loc in range(len(shape_obj.points)):
        longs[loc] = shape_obj.points[loc][0]
        lats[loc] = shape_obj.points[loc][1]
        
    plt.plot(longs, lats, color='grey')
    x0 = np.mean(longs)
    y0 = np.mean(lats)
    
    ax.set_xlim(shape_obj.bbox[0] - 0.05, shape_obj.bbox[2] + 0.05)
    state = sf.records()[dist_id][0]
    dist_num = sf.records()[dist_id][2]
    
    if dist_num[-1] == '1':
        oral_ind = 'st'
    elif str(dist_num)[-1] == '2':
        oral_ind = 'nd'
    elif str(dist_num)[-1] == '3':
        oral_ind = 'rd'
    else:
        oral_ind = 'th'
    
    ax.set_title('%s\'s %s%s District' % (state, dist_num, oral_ind))
    ax.fill(longs, lats, fill_col)
    
    return x0, y0
```

```python
def plot_multiple_shapes(sf, dist_ids, fill_cols, figsize=(10, 6), title='<>', xlim=None, ylim=None):
    '''
        Plots multiple shapes
        
        sf: Shapefile object containing plotting data
        dist_ids: array of shapefile id's of districts we wish to plot
        fill_cols: array of colors that we wish to fill each shape with
    '''
    
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_aspect('equal')
    
    dist_nums = []
    
    for i in range(len(dist_ids)):
        shape_obj = sf.shape(dist_ids[i])
    
        ## Plot longs on the x axis, lats on the y axis
        longs = np.zeros((len(shape_obj.points), 1))
        lats = np.zeros((len(shape_obj.points),1))
    
        for loc in range(len(shape_obj.points)):
            longs[loc] = shape_obj.points[loc][0]
            lats[loc] = shape_obj.points[loc][1]
        
        plt.plot(longs, lats, color='black')
        x0 = np.mean(longs)
        y0 = np.mean(lats)
    
        ax.fill(longs, lats, fill_cols[i])
    
        state = sf.records()[dist_ids[i]][0]
        dist_num = sf.records()[dist_ids[i]][2]
    
        if dist_num[-1] == '1':
            oral_ind = 'st'
        elif str(dist_num)[-1] == '2':
            oral_ind = 'nd'
        elif str(dist_num)[-1] == '3':
            oral_ind = 'rd'
        else:
            oral_ind = 'th'
            
        dist_nums.append(dist_num + oral_ind)
     
    if title == '<>':
        title = state + '\'s ' + ', '.join(dist_nums) + ' Districts'
    
    ax.set_title(title)
    
    cmap_r = mpl.colors.ListedColormap(['#F44336', '#E53935', '#E53935', '#E53935', '#E53935', '#E53935', '#E53935'], )
    cmap_r.set_over('0.25')
    cmap_r.set_under('0.25')
    
    #cmap_d = mpl.colors.ListedColormap(['#2196F3', '#1E88E5', '#1E88E5', '#1E88E5', '#1E88E5', '#1E88E5', '#1E88E5'])
    
    bounds = [1, 3, 5, 7, 9, 15]
    norm = mpl.colors.BoundaryNorm(bounds, cmap_r.N)
    
    fig.colorbar(mpl.cm.ScalarMappable(cmap=cmap_r, norm=norm))
    
    if xlim != None:
        ax.set_xlim(xlim)
    if ylim != None:
        ax.set_ylim(ylim)
        
    
    
    return x0, y0
```

```python
house_110 = pd.read_csv('../congress-data/house/house_110.csv', index_col=0)
```

```python
house_110['pct'] = house_110['missed_votes'] / house_110['total_votes']
```

```python
california = house_110[house_110['state'] == 'CA']
```

```python
california['full_state'] = 'California'
```

```python
party_list = california['party'].values
pct_list = california['pct'].values
```

```python
california_fill_cols = map_colors(party_list, pct_list)
```

```python
df['INDEX'] = df.index
sf_indices = df.merge(california, right_on=['full_state', 'district'], left_on=['STATENAME', 'DISTRICT'])['INDEX']
```

```python
plot_multiple_shapes(sf, sf_indices, california_fill_cols, 
                     title='California\'s Districts',
                     xlim=(-123, -120.5), 
                     ylim=(36, 39))
```
